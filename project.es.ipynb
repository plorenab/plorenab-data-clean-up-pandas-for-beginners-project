#### Ejercicio 01. ¿Cuál es la casa más cara en todo el dataset? (★☆☆)

Imprime la dirección y el precio de la casa seleccionada. Por ejemplo:

`La casa con dirección en Calle del Prado, Nº20 es la más cara y su precio es de 5000000 USD`

```python
# Ordenamos la tabla en orden descendente para 'price' y nos quedamos con la primera posición (vivienda mas cara)
sorted = ds.sort_values('price', ascending= False)
most_expensive = sorted.iloc[0]
print(f'La casa con dirección en {most_expensive["address"]} es la más cara y su precio es de {most_expensive["price"]} USD')
```

#### Ejercicio 02. ¿Cuál es la casa más barata del dataset? (★☆☆)

Este ejercicio es similar al ejercicio anterior, solo que ahora buscamos la casa con el precio más bajo. Recuerda imprimir la dirección y el precio de la casa seleccionada. Por ejemplo:

`La casa con dirección en Calle Alcalá, Nº58 es la más barata y su precio es de 12000 USD`

```python
ds_sin_ceros = ds[ds["price"]!= 0] #Esta línea deja fuera del dataset las viviendas donde el precio es 0 

# Ordenamos la tabla en orden ascendente para 'price'
sorted = ds_sin_ceros.sort_values('price', ascending= True)
# Nos quedamos con la primera posición de la tabla (vivienda mas barata y precio distinto de 0)
cheapest = sorted.iloc[0]

print(f'La casa con dirección en {cheapest["address"]} es la más barata y su precio es de {cheapest["price"]} USD')
```

#### Ejercicio 03. ¿Cuál es la casa más grande y la más pequeña del dataset? (★☆☆)

Imprime la dirección y el área de las casas seleccionadas. Por ejemplo:

`La casa más grande está ubicada en Calle Gran Vía, Nº38 y su superficie es de 5000 metros`

`La casa más pequeña está ubicada en Calle Mayor, Nº12 y su superficie es de 200 metros`

Este ejercicio es similar al anterior, pero buscamos las casas más grandes y pequeñas en base a su superficie.

```python
# Para encontrar la casa mas GRANDE, ordenamos la tabla en orden descendente para 'surface'
sorted_big = ds.sort_values('surface', ascending= False)
# Nos quedamos con la primera posición de la tabla (vivienda mas grande)
bigger_house = sorted_big.iloc[0]

print(f'La casa más grande está ubicada en {bigger_house["address"]} y su superficie es de {bigger_house["surface"]} metros')


# Para encontrar la casa mas PEQUEÑA, ordenamos la tabla en orden ascendente para 'surface'
sorted_small = ds.sort_values('surface', ascending= True)
# Nos quedamos con la primera posición de la tabla (vivienda mas pequeña)
smaller_house = sorted_small.iloc[0]

print(f'La casa más pequeña está ubicada en {smaller_house["address"]} y su superficie es de {smaller_house["surface"]} metros')
```

#### Ejercicio 04. ¿Cuantas poblaciones contiene el dataset? (★☆☆)

Cuenta el número de poblaciones únicas en la columna 'level5' e imprime el nombre de las poblaciones separadas por coma. Por ejemplo:

`> print(populations)`

`population1, population2, population3, ...`

```python
# Eliminamos los valores repetidos de la columna 'level5' y deja solo una copia de cada población
populations = ds["level5"].drop_duplicates()
# Contamos el numero de poblaciones diferentes que contiene el dataset
population_number = populations.count()
# Unimos todas las poblaciones en una sola cadena de texto, separados por comas y un espacio
comma_populations = ', '.join(populations)

print(f'El dataset contiene {population_number} poblaciones diferentes. Estas poblaciones son: {comma_populations}')
```

#### Ejercicio 05. ¿El dataset contiene valores nulos (NAs)? (★☆☆)

Imprima un booleano (`True` o `False`) para verificar si hay valores nulos seguido de las columnas que contiene el NAs.

```python
# Devuelve True si al menos hay un valor nulo en ds. El primer any() devuelve bool por columnas
hay_nulo = ds.isnull().any().any() 
print("El dataset contiene valores nulos?", hay_nulo)

# Guardamos el nombre de las columnas que tengan al menos un valor Nan
null_columns = ds.columns[ds.isnull().any()]

print("Columnas con valores nulos:", ", ".join(null_columns))
```

#### Ejercicio 06. Elimina los valores nulos (NAs) del dataset, si aplica (★★☆)

Despues de eliminar los valores nulos, compara el tamaño del DataFrame antes y después de la eliminación.

#### Ejercicio 07. ¿Cuál es la media de precios en la población de "Arroyomolinos (Madrid)"? (★★☆)

Imprima el valor obtenido de la columna level5.

```python
#Filtramos el dataset para mantener solo las filas donde level5 = Arroyomolinos
poblacion = ds[ds["level5"] == "Arroyomolinos (Madrid)"] 

#Hago la media de precios de mi población
media_precios = poblacion["price"].mean() 
print("La media de precio en Arroyomolinos (Madrid) es", round(media_precios, 2), "USD")
```

#### Ejercicio 08. Trazar el histograma de los precios para la población de "Arroyomolinos (Madrid)" y explica qué observas (★★☆)

Imprime el histograma de los precios y escribe en la celda del Markdown un breve análisis del trazado.


```python
import matplotlib.pyplot as plt

#Elaboramos un histograma de precios a partir de la variable 'población' creada en el ejercicio anterior
histograma = plt.hist(poblacion['price'])
print(histograma)

```
